Having computed factor loadings for eighteen factors, we can analyze the results, hopefully giving them a physical interpretation.

```python
import pandas as pd
loadings = pd.read_csv('factor_loadings_feb4.csv', index_col=0)
pd.set_option('display.max_columns', None)
```

```python
def style(df):
    return df.style.background_gradient(axis='columns', cmap = 'RdYlGn', vmin=-1, vmax=1)
style(loadings)
```

We should give each factor a descriptive name. Let's start off by referring to them by the column with the greatest loading.

```python
loadings.insert(0, 'name',  loadings.idxmax(axis='columns'))
style(loadings)
```

0, 2, and 4 all relate to angles. Let's investigate closer:

```python
style(loadings.iloc[[0,2, 4]][[col for col in loadings.columns if 'ang' in col]])
```

It looks like these reach refer to the change in angle -- there are three of them, because there are three *independent* windows. The overlaps are highly correlated.

Actually, this seems to indicate a gneral pattern: all of these factors refer to something which was computed across windows. This suggests that breaking it up by window is not appropriate. if we wish to have a small covariance -- by definition, the motion will be highly correlated.
